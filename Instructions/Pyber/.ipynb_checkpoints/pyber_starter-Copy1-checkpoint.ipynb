```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read the City and Ride Data
city_df = pd.read_csv("data/city_data.csv")
ride_df = pd.read_csv("data/ride_data.csv")
#city_df.head()
#ride_df.head()

# Combine the data into a single dataset
city_ride_df = pd.merge(ride_df, city_df, on='city')
# Display the data table for preview
city_ride_df.head()

#group data by City Type
group_by_type_df = city_ride_df.groupby(['type'])

#Extract Rural, Urban and Suburban data into three seprate DataFrames 
rural_df = group_by_type_df.get_group('Rural')
urban_df = group_by_type_df.get_group('Urban')
suburban_df = group_by_type_df.get_group('Suburban')

rural_city_df = rural_df.groupby(['city'])
urban_city_df = urban_df.groupby(['city'])
suburban_city_df = suburban_df.groupby(['city'])

#find the average of fare by city
avg_rural_city_fair = rural_city_df['fare'].mean()
avg_urban_city_fair = urban_city_df['fare'].mean()
avg_suburban_city_fair = suburban_city_df['fare'].mean()

#find the count of rides by city
total_rural_city_ride = rural_city_df['ride_id'].count()
total_urban_city_ride = urban_city_df['ride_id'].count()
total_suburban_city_ride = suburban_city_df['ride_id'].count()

#find the count of drivers by city
total_rural_city_drivers = rural_city_df['driver_count'].count()
total_urban_city_drivers = urban_city_df['driver_count'].count()
total_suburban_city_drivers = suburban_city_df['driver_count'].count()

#Define the figure size
fig=plt.figure(figsize=(10,10))

#edit the chart parameters below 
wedges=plt.scatter(total_rural_city_ride, avg_rural_city_fair,s=total_rural_city_drivers*15
                   , c="Gold", alpha=0.5
                   ,marker='o'
                   ,facecolors=["Gold"]
                   ,edgecolors=["Black"]
                   ,label="Rural") 
wedges.set(edgecolor='blue',linewidth=1)

plt.scatter(total_urban_city_ride, avg_urban_city_fair, s=total_urban_city_drivers*15, c="Lightskyblue", alpha=0.5,
           marker='o',facecolors=["Lightcoral"],edgecolors=["Red"],label="Urban")
plt.scatter(total_suburban_city_ride, avg_suburban_city_fair, s=total_suburban_city_drivers*15, c="Coral", alpha=0.5,
           marker='o',facecolors=["Lightskyblue"],edgecolors=["blue"],label="Suburban")



plt.title("Pyber ride sharing data(2016)")
#Set axes limit
plt.xlim(0,40)
plt.ylim(19,45)

#Set axes Labels
plt.xlabel("Total number of rides(per city)")
plt.ylabel("Average fare ($)")

#Set legend location to upper right
plt.legend(loc='upper right')
plt.grid()
plt.text(45, 42, "Note:", fontsize=12)
plt.text(45, 40, "Circle size corelates with driver count per city", fontsize=12)
plt.savefig('pyber_ride_sharing_data.png')
plt.show()
```

#Total Fares by City Type

```python
# % of Total Fares by City Type
total_fares_df = city_ride_df.groupby('type')['fare'].sum()
print(total_fares_df)

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
sizes   = total_fares_df
explode = (0, 0, 0.1)  # only "explode" the 2nd slice
labels  = 'Rural', 'Suburban','Urban'
colors  = ("Gold","Lightskyblue","Coral")

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%',shadow=True,
        startangle=90,counterclock=False)
        
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title("% Total Fares by City type")
plt.savefig('pyber_fares_by_city_type.png')
plt.show()

```

```python
# % of Total Rides by City Type
total_rides_df = city_ride_df.groupby('type')['ride_id'].count()
print(total_rides_df)

sizes   = total_rides_df
explode = (0, 0, 0.1)  
labels  = 'Rural', 'Suburban','Urban'
colors  = ("Gold","Lightskyblue","Coral")


fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title("% of Total Rides by City Type")
plt.savefig('pyber_rides_by_city_type.png')
plt.show()
```

```python
#  % of Total Drivers by City Type

total_drivers_df = city_df.groupby('type')['driver_count'].sum()


print(total_drivers_df)


# Pie chart, where the slices will be ordered and plotted counter-clockwise:
sizes   = total_drivers_df
explode = (0, 0, 0.1)  # only "explode" the 2nd slice
labels  = 'Rural', 'Suburban','Urban'
colors  = ("Gold","Lightskyblue","Coral")

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title("% of Total Drivers by City Type")
plt.savefig('pyber_drivars_by_city_type.png')
plt.show()

```
